### AB steps


A complete A/B test typically includes the following components:

1. Analyze the Current Situation and Formulate Hypotheses: Analyze the business to identify the highest priority improvement areas, formulate hypotheses, and propose optimization suggestions.

2. Set Metrics: Establish primary metrics to measure the effectiveness of each version and set secondary metrics to evaluate other impacts.

3. Design and Development: Design the prototype of the optimized version and complete its development.

4. Determine Test Duration: Decide on the duration for which the test will run.

5. Determine Splitting Strategy: Decide the splitting ratio for each test version and other splitting details.

6. Collect and Analyze Data: Gather experimental data and evaluate the effectiveness and impact.

7. Draw Conclusions:
Determine to release the new version.
Adjust the splitting ratio and continue testing.
Optimize the iterative plan, redevelop, and return to step 1.

#### Experimental Steps


Hypotheses:

- Check normality
- If normally distributed, check homogeneity of variance

Steps:

- Split and define control and test groups
- Apply the Shapiro-Wilk test to check for normality
- If assumptions are met, apply Levene's test to check for homogeneity of variance
- If assumptions are met and variances are homogeneous, apply the T-test
- If assumptions are met but variances are not homogeneous, apply the Welch's test
- If assumptions are not met, directly apply the Mann-Whitney U test



#### Considerations

<font color = red>Points to Note</font>

- Test Duration: The duration of the test should not be too short; otherwise, the users participating in the experiment will mostly be high-frequency users of the product.

- Splitting (or Sampling): Ensure simultaneity, homogeneity, uniqueness, and uniformity.

1. Simultaneity: The split should be simultaneous, and the test should also be conducted simultaneously.

2. Homogeneity: Also known as similarity, requires that the user groups split are similar in characteristics across various dimensions. Grouping can be based on device characteristics (e.g., phone model, operating system version, phone language) and other user tags (e.g., gender, age, new vs. old users, membership level). Each A/B test can select specific user groups for testing.
Consideration: How to determine if the groups are truly homogeneous? An AAB test can be used. Extract two sets of traffic for A version testing, perform an AA test, and conduct AB testing with the B version. By examining whether there is a significant difference between groups A1 and A2, you can determine if the experimental split is homogeneous.
3. Uniqueness: Ensure that users are not counted in the test more than once.

4. Uniformity: Requires that traffic in each group is uniform. This is often managed by specialized A/B testing tools. There is also an article that describes an implementation in Python, which involves adding a salt value to the user ID, hashing it, and calculating a float between 0 and 1. #### Why determine if it is normal?

Determining whether data follows a normal distribution is important because many statistical methods (such as the t-test) are more accurate and reliable when the data is normally distributed. Normality testing is used to confirm whether the chosen statistical methods are suitable for the data to avoid drawing incorrect conclusions in hypothesis testing.

If the data follows a normal distribution, parametric statistical methods (such as the t-test) can be used for hypothesis testing, allowing for more accurate estimation of differences in population parameters (such as the population mean). However, if the data does not follow a normal distribution, using parametric methods may lead to bias or incorrect conclusions.

In A/B testing, we typically use normality testing to verify whether the statistical methods we are using are appropriate for the data, ensuring that our conclusions are reliable and accurate. If the data does not follow a normal distribution, we can opt to use non-parametric statistical methods (such as the Mann-Whitney U test) for hypothesis testing. These methods do not make any assumptions about the distribution of the data, making them more flexible and applicable to different types of data.



#### What if the data is not normal?

Common non-parametric statistical methods include:

1. Mann-Whitney U test: Used to compare the median difference between two independent samples. (This is generally used.)
2. Wilcoxon signed-rank test: Used to compare the median difference between two related samples.
3. Kruskal-Wallis test: Used to compare the median difference among multiple independent samples.
4. Friedman test: Used to compare the median difference among multiple related samples.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import shapiro, ttest_ind, mannwhitneyu, levene
import statsmodels.stats.api as sms

def AB_Test(dataframe, group, target, alpha=0.05, visualize=True):
    #Split A/B Groups

    groupA = dataframe[dataframe[group] == "A"][target]
    groupB = dataframe[dataframe[group] == "B"][target]

    #  perform hypothesis testing for normality
    ntA = shapiro(groupA)[1] < alpha
    ntB = shapiro(groupB)[1] < alpha

    # parametric tests
    leveneTest = levene(groupA, groupB)[1] < alpha

    if (ntA == False) & (ntB == False):
        # homogeneity of variance
        if leveneTest == False:
            ttest, p_value = ttest_ind(groupA, groupB, equal_var=True)
            test_type = "Parametric"
        else:
            ttest, p_value = ttest_ind(groupA, groupB, equal_var=False)
            test_type = "Parametric (Unequal Variance)"
    else:
        # non-parametric tests
        ttest, p_value = mannwhitneyu(groupA, groupB, alternative='two-sided')
        test_type = "Non-Parametric"

    # calculate effect size
    mean_diff = groupB.mean() - groupA.mean()


    nobs_A = len(groupA)
    nobs_B = len(groupB)
    pooled_std = np.sqrt(((nobs_A-1)*groupA.var() + (nobs_B-1)*groupB.var()) / (nobs_A + nobs_B - 2))
    effect_size = mean_diff / pooled_std

    # calculate power
    observed_power = sms.tt_ind_solve_power(effect_size=effect_size, nobs1=nobs_A, alpha=alpha, alternative='two-sided')

    # determine if the test is significant
    ab_hypothesis = p_value < alpha
    comment = "A/B groups are similar！" if not ab_hypothesis else "A/B groups are not similar！"

    result = pd.DataFrame({
        "Test Type": [test_type],
        "target": [target],
        "Homogeneity": ["Yes" if not leveneTest else "No"],
        "AB Hypothesis": ["Reject H0" if ab_hypothesis else "Fail to Reject H0"],
        "p-value": [p_value],
        "Mean Difference": [mean_diff],
        "Effect Size": [effect_size],
        "Observed Power": [observed_power],
        "Comment": [comment]
    })

    #
    print("# A/B Test Hypotheses")
    print("H0: A == B")
    print("H1: A != B", "\n")


    print(f"if meets the assumption of normality？\nAgroup: {not ntA}, Bgroup: {not ntB}\n")
    print(f"whether the assumption of homogeneity of variance: {not leveneTest}", "\n")


    if visualize:
        plt.figure(figsize=(6, 4))
        sns.histplot(data=dataframe, x=target, hue=group, kde=True)
        plt.axvline(groupA.mean(), color='blue', linestyle='dashed', linewidth=2, label=f'Mean A: {groupA.mean():.2f}')
        plt.axvline(groupB.mean(), color='orange', linestyle='dashed', linewidth=2, label=f'Mean B: {groupB.mean():.2f}')
        plt.title('Distribution of Groups')
        plt.legend()
        plt.show()

    return result


result = AB_Test(dataframe=ab, group="version", target="sum_gamerounds")

result
```

```python
from statsmodels.stats.proportion import proportions_ztest


# dt must be in the form of a DataFrame with four columns: metrics, group, population, and target, representing the variable name, group, sample size (the denominator of the retention rate), and hit sample size (the numerator of the retention rate), respectively.
#base is the baseline group, which is selected from the group column. All other groups will be compared with this group using a two-sample T-test.
#The way to set base: base = "name of the group
def binary_T_test(dt=None, base_group=None):

    dt.columns = ["metrics", "group", "population", "target"]

    # Get the unique metrics and group names
    metrics_vec = dt['metrics'].unique()
    group_vec = dt['group'].unique()


    # metrics	 group population	target
    # retention_1	A	 44699	  20034
    # retention_1	B	 45489	  20119
    # retention_7	A	 44699	  8501
    # retention_7	B	 45489	  8279


    # Initialize a DataFrame to store the p-values
    p_values = pd.DataFrame(index=metrics_vec, columns=group_vec)

    # Loop through each metric
    for i, metric in enumerate(metrics_vec):
        # Get the data for the current metric
        sub_tt = dt[dt['metrics'] == metric]

        # Initialize an array to store the p-values for each group
        p_value = np.zeros(len(group_vec))

        # Loop through each group
        for j, group in enumerate(group_vec):
            # Get the population and target size for the current group
            pop0 = sub_tt.loc[sub_tt['group'] == base_group, 'population'].values[0]
            tar0 = sub_tt.loc[sub_tt['group'] == base_group, 'target'].values[0]

            # Get the population and target size for the current group
            pop1 = sub_tt.loc[sub_tt['group'] == group, 'population'].values[0]
            tar1 = sub_tt.loc[sub_tt['group'] == group, 'target'].values[0]

            # Calculate the count and nobs for the current group
            count = np.array([tar0, tar1])
            nobs = np.array([pop0, pop1])

            # Perform a two-sample T-test
            _, p_value[j] = proportions_ztest(count, nobs, alternative='two-sided')

        # Store the p-values for the current metric
        p_values.loc[metric, :] = p_value

    # Return the p-values
    results = {'P_values': p_values}
    return results


# Call the function with Python DataFrame and base group name
results = binary_T_test(dt=rent_res, base_group="A")
print(results['P_values'])
```
